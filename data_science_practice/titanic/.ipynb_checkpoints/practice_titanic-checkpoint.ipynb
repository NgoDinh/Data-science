```python
# Load data and set up package
#----------------------------------------#

import re
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelBinarizer
from sklearn.feature_selection import VarianceThreshold
from sklearn.model_selection import GridSearchCV, train_test_split, cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
%matplotlib inline

df = pd.read_csv('train.csv')
df.head(10)
```

```python
# Load data and set up package
#----------------------------------------#

def feature_exp(data):
    features_name = data.keys()
    for feature in features_name:
        total_na = data[[feature]].isnull().sum()
        print((feature, total_na))

def find_label(data):
    features_name = data.keys()
    final_dict = {}
    for feature in features_name:
        set_labels = set(data[feature])
        number_labels = len(set_labels)
        if number_labels < 10:
            final_dict[feature] = set_labels
    return final_dict

print(feature_exp(df))
print('-'*10)
print(find_label(df))
```

```python
# Check most frequency value in 'Embarked' feature
#----------------------------------------#

df['Embarked'].value_counts()
```

```python
# Feature construction
#----------------------------------------#

def type_of_ticket(ticket):
    have_text = re.search('[A-Za-z]', ticket)
    if have_text:
        return 'T' # T: have Test
    else:
        return 'I' # I: just integer

def feature_engineer(data):
    Sex_dict = {'male': 0, 'female': 1}
    Embarked_dict = {'C': 0, 'S': 1, 'Q': 2}
    Ticket_dict = {'T': 0, 'I': 1}
    drop_elements = ['PassengerId', 'Name', 'Cabin', 'Ticket']

    data['Embarked'] = data['Embarked'].fillna('S')
    data['Embarked'] = data['Embarked'].map(Embarked_dic).astype(int)
    # Encoding 'Embarked' feature

    data['Has_Cabin'] = data['Cabin'].apply(lambda x: x[0] if type(x) == str else 'No')
    # Check cabin of Passenger, if NA return 'No'
    encoder = LabelBinarizer()
    Has_Cabin_Encode = encoder.fit_transform(data['Has_Cabin'])
    Has_Cabin_Encode = pd.DataFrame(Has_Cabin_Encode, columns=encoder.classes_)
    data = pd.concat([data, Has_Cabin_Encode], axis=1)
    data = data.drop('Has_Cabin', axis=1)
    # Encoding new feature 'Has_cabin' with LabelBinarizer

    data['Family'] = data['SibSp'] + data['Parch'] 
    # Create new feature to caculate total family member of Passenger

    data['Type_of_ticket'] = data['Ticket'].apply(type_of_ticket)
    data['Type_of_ticket'] = data['Type_of_ticket'].map(Ticket_dict).astype(int)
    # Split 'Ticket' feature into two group: just have number | number and test

    data['Sex'] = data['Sex'].map(Sex_dict).astype(int)
    # Encoding 'Sex' feature

    data['Age'][np.isnan(data['Age'])] = data['Age'].median()
    # Fill NA in 'Age' feature with mean of this column

    data = data.drop(drop_elements, axis = 1)
    # drop unecessary features
    
    return data
    
df = feature_engineer(df)
```

```python
# Feature selection
#----------------------------------------#

sel = VarianceThreshold(threshold=.8 * (1 - .8))
sel.fit_transform(df)
# Remove the feature that the values in it are relatively homogeneous.

colormap = plt.cm.RdBu
plt.figure(figsize=(14,12))
plt.title('Pearson Correlation of Features', y=1.05, size=15)
sns.heatmap(df.astype(float).corr(),linewidths=0.1,vmax=1.0, 
            square=True, cmap=colormap, linecolor='white', annot=True)
# find correlation between features
```

```python
# Use gridsearch to find best parameter for tree model
#----------------------------------------#

X = df.loc[:, df.columns != 'Survived']
y = df['Survived']

grid = {'criterion': ['gini', 'entropy'],
       'max_depth': [5, 10, 15],
       'min_samples_split': [2, 10, 50, 100],
       'min_weight_fraction_leaf' : [0.001, 0.01, 0.05]}

search = GridSearchCV(DecisionTreeClassifier(), grid, scoring = 'accuracy', cv = 5, refit = True)
search.fit(X, y)
search.best_params_
```

```python
# train and evaluate score with best parameter
#----------------------------------------#

best_tree = search.best_estimator_
best_scores = cross_val_score(best_tree, X, y, scoring = 'accuracy', cv = 5)
best_scores.mean(), best_scores.std()
```
