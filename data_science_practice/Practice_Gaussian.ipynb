```python
#Prepare data
#==========================================#

import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris

data = load_iris().data
target = load_iris().target

target_2 = np.where(target == 2)
data_01 = np.delete(data, target_2, axis=0)
target_01 = np.delete(target, target_2, axis=0)
data_01 = data_01[:,[0,1]]
```

```python
#Build GDA model
#==========================================#

class GDA(object):
    def __init__(self, data):
        self.data = data
        
    def phi(self):
        m = self.data.shape[0]
        return len(np.where(self.data[:,-1] == 1)[0])/m
    
    def miu_0(self):
        target_0_index = np.where(self.data[:,-1] == 0)
        sum_by_cols = np.sum(self.data[target_0_index,:-1], axis=1)
        return sum_by_cols / len(np.where(self.data[:,-1] == 0)[0])

    def miu_1(self):
        target_1_index = np.where(self.data[:,-1] == 1)
        sum_by_cols = np.sum(self.data[target_1_index,:-1], axis=1)
        return sum_by_cols / len(np.where(self.data[:,-1] == 1)[0])
    
    def sigma(self):
        temp_matrix = []
        m = self.data.shape[0]
        miu_0_value = self.miu_0()
        miu_1_value = self.miu_1()
        for i in range(m):
            if self.data[i,-1] == 0:
                temp_matrix.append((self.data[i,:-1] - miu_0_value)[0].tolist())           
            else:
                temp_matrix.append((self.data[i,:-1] - miu_1_value)[0].tolist())
        temp_matrix = np.array(temp_matrix)
        return (temp_matrix.T.dot(temp_matrix)/m)
    
#     def sigma_new(self):
#         m = self.data.shape[0]
#         n = self.data.shape[1] - 1
#         sum_matrix = np.zeros((2,2))
#         miu_0_value = self.miu_0()
#         miu_1_value = self.miu_1()
#         for i in range(m):
#             if self.data[i,-1] == 0:
#                 temp = self.data[i,:-1] - miu_0_value
#                 sum_matrix = sum_matrix + temp.T.dot(temp)
#             else:
#                 temp = self.data[i,:-1] - miu_1_value
#                 sum_matrix = sum_matrix + temp.T.dot(temp)
#         return (sum_matrix/m)
        
    def caculate_pxy(self, new_data):
        n = self.data.shape[1] - 1
        sigma = self.sigma()
        phi = self.phi()
        X = new_data
        X_0 = X - self.miu_0()
        X_1 = X - self.miu_1()
        const = 1 / ((2 * np.pi) ** (n / 2) * np.linalg.det(sigma) ** 0.5)
        py_0 = const * np.exp(-0.5 * (X_0.dot(np.linalg.inv(sigma))).dot(X_0.T)) * (1 - phi)
        py_1 = const * np.exp(-0.5 * (X_1.dot(np.linalg.inv(sigma))).dot(X_1.T)) * phi
        # If y == 0 : P(y) = (1 - phi) else: P(y) = phi
        if py_1 > py_0:
            print('New data:{} blong to class 1'.format(X))
            print('-'*5)
        else:
            print('New data:{} blong to class 0'.format(X))
            print('-'*5)
        print('P_0:{}, P_1:{}'.format(py_0, py_1))
        print('='*10)
        return(py_0, py_1)
```

```python
#Run model model
#==========================================#

data_final = np.c_[data_01, target_01]
gda = GDA(data_final)
gda.caculate_pxy([4.5, 3.5])
gda.caculate_pxy([6., 2.5])
gda.caculate_pxy([5.5, 3.])
```
